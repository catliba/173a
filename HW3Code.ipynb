```python
import numpy as np
import pandas as pd
```

```python
filePath = '/Users/caleb/Downloads/HW2_ellipse.csv'
df = pd.read_csv(filePath, header = None)
df.plot.scatter(x = 0, y = 1)
```

```python
A = df.to_numpy()
A = A**2
X = np.dot(A.T, A)
mu = 1/(2*(np.linalg.norm(X,2)))
print ('mu = ', mu)
a = np.array([[1],[1]])
b = np.ones((1000,1))
stop_condition = 100

for i in range(stop_condition):
    gradient = 2 * np.dot(A.T,np.dot(A,a) - b)
    a = a - mu * gradient

print('a = ', a)
func_a = np.linalg.norm(np.dot(A,a)-b)**2
fa = print('func_a = ', func_a)
```
